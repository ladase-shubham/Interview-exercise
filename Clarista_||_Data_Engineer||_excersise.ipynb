**Problem-2**

```python
price_adj_df_cp = price_adj_df[['date','symbol','close']]
```

```python
price_adj_df_cp['Latest Date'] = pd.to_datetime(price_adj_df_cp['date'])
price_adj_df_cp.sort_values(['Latest Date'],ascending=False).groupby('symbol')
# price_adj_df_cp[price_adj_df_cp['symbol']=='ABT']
list_of_dataframes = []
for symbol in price_adj_df_cp['symbol'].unique():
  temp_df = price_adj_df_cp[price_adj_df_cp['symbol']==symbol]
  temp_df['Close Price on Latest Date -1 day'] = temp_df['close'].shift(-1)
  temp_df['% Change'] = (((temp_df['close'] - temp_df['Close Price on Latest Date -1 day']) *100) / temp_df['Close Price on Latest Date -1 day'])
  # temp_df.rename(columns = {'symbol':'Ticker'}, inplace = True)
  list_of_dataframes.append(temp_df)
  del temp_df
new_df = pd.concat(list_of_dataframes)

```

```python
print('Before -->',len(new_df))
# join dataframe
final_df = new_df.join(securities_df.set_index('Ticker symbol'), on='symbol')
print('After -->',len(final_df))
```

**Problem-3**

```python
samples_symbols = price_adj_df_cp['symbol'].sample(5).to_list()
```

```python
samples_symbols #--['XLNX', 'ICE', 'HBI', 'DHI', 'EIX']
df_problem_2 = price_adj_df_cp
df_problem_2['Latest Date'] = pd.to_datetime(df_problem_2['date'])
df_problem_2.sort_values(['Latest Date'],ascending=False).groupby('symbol')
list_of_dataframes = []
for symbol in samples_symbols:
  temp_df = df_problem_2[df_problem_2['symbol']==symbol]
  temp_df['5 Day Rolling Mean'] = temp_df.close.rolling(5).mean()
  list_of_dataframes.append(temp_df)
  del temp_df
filtered_df = pd.concat(list_of_dataframes)
filtered_df.head(10)
```

```python
# required columns
# a. Symbol
# b. Date
# c. Close Price
# d. 5 Day Rolling Mean
# !pip install xlsxwriter
# filtered_df.rename(columns = {'symbol':'Symbol', 'date': 'Date', 'close': 'Close Price'}, inplace = True)
from openpyxl import load_workbook
path = "/drive/MyDrive/InterviewSubmission/Copy of submission.xlsx"
book = load_workbook(path)
writer = pd.ExcelWriter(path, engine = 'openpyxl')
writer.book = book
filtered_df[['Symbol', 'Date', 'Close Price', '5 Day Rolling Mean']].to_excel(writer, sheet_name='Rolling mean')
```
